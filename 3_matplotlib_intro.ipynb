```python
plt.plot(a,a+5)   # x is array a and y is array a+5
plt.show()
```

```python
plt.figure(figsize=(10,8))   # to increase size of graph

plt.plot(a,a,label=" x square",color="yellow")
plt.plot(a,a2,label="x",color="red",marker="*")
plt.legend()  #to define top left box of label

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("plot graph")
plt.show()
```

```python
# now bar graph
#in a function press shift + tab once,twice , thrice to see what that function does
x=np.array([1,2,3,4,5])
y1=np.array([10,13,9,23,5])
y2=np.array([34,21,67,9,4])
plt.bar(x+0.3,y1,width=0.3,color="red",label="sales in 2018")  # x+.3 to avoid overlapping
plt.bar(x,y2,width=0.3,color="blue",label="sales in 2019")
plt.legend()
plt.show()
```

```python
# WORKING WITH IMAGES
img=plt.imread("pokemon.jpg")
print(img.shape)
plt.imshow(img)
plt.show()
```

```python
box=np.zeros((100,100,3))      # note the spelling of zeros   , the 3 is for R B G coor component
plt.imshow(box)
plt.show()
```

```python
box=np.zeros((100,100,3),dtype='uint8')
box[:,:,0]=100  # Red
box[:,:,1]=40    # Green
box[:,:,2]=100    # blue
plt.imshow(box)
plt.show()
```

```python
#stack images nowv
v=np.vstack((img,img))
plt.imshow(v)
border=np.zeros((v.shape[0],400,3),dtype='uint8')+255    # create a border of white pixels, row len same as that of img row len 
h=np.hstack((v,border,v))                                # and number of cols as the length of the white pixel border and 3 bcoz 
plt.imshow(h)                                            # there are 3 colors , compulsory to be written like this bcoz we are inserting an image and not just anything
```
